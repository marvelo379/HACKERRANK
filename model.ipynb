```python
def age(x):
  if pd.isnull(x):
    return -1
  num=int(x.split(' ')[0])  
  if 'year' in x:
    return num*365
  elif 'month' in x:
    return num*30
  elif 'week' in x:
    return num*7
  else:
    return num
alldata['age_upon_outcome']=alldata['age_upon_outcome'].apply(age)
alldata          
```
